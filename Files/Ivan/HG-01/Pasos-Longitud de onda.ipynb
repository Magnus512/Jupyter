```python
cb = np.array(AmplitudD)
indices = peakutils.indexes(cb, thres=0.02/max(cb), min_dist=0.1)

trace = go.Scatter(
    x=pasos,
    y=AmplitudD,
    mode='lines',
    name='Original Plot'
)

trace2 = go.Scatter(
    x=indices+2501,
    y=[AmplitudD[j] for j in indices],
    mode='markers',
    marker=dict(
        size=8,
        color='rgb(255,0,0)',
        symbol='cross'
    ),
    name='Detected Peaks'
)

data = [trace, trace2]
py.iplot(data, filename='Amplitud-Pasos')
```
