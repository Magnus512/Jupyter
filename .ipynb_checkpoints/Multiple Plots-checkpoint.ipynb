```python
from matplotlib.pyplot import figure, show
import matplotlib.pyplot as plt
from numpy import arange, sin, pi

t = arange(0.0, 10, 0.01)

fig = figure(1,figsize=(10,20))
grid = plt.GridSpec(4, 1, hspace=0.4, wspace=0.2)

laser1 = fig.add_subplot(grid[0,0])
laser1.plot(t, sin(2*pi*t))
laser1.grid(True)
laser1.set_ylabel('Amplitud').set_fontsize('16')
laser1.set_title('Señar Roja (660nm)').set_fontsize('16')
laser1.set_xlabel('Tiempo').set_fontsize('16')
laser1.set_xlim((0,1))

laser2 = fig.add_subplot(grid[1,0])
laser2.plot(t, sin(2*2*pi*t))
laser2.grid(True)
laser2.set_ylabel('Amplitud').set_fontsize('16')
laser2.set_title('Señar Infrarroja (780nm)').set_fontsize('16')
laser2.set_xlabel('Tiempo').set_fontsize('16')

laser3 = fig.add_subplot(grid[2,0])
laser3.plot(t, sin(3*2*pi*t))
laser3.grid(True)
laser3.set_ylabel('Amplitud').set_fontsize('16')
laser3.set_title('Señar Infrarroja (808nm)').set_fontsize('16')
laser3.set_xlabel('Tiempo').set_fontsize('16')

laser4 = fig.add_subplot(grid[3,0])
laser4.plot(t, sin(4*2*pi*t))
laser4.grid(True)
laser4.set_ylabel('Amplitud').set_fontsize('16')
laser4.set_title('Señar Infrarroja (858nm)').set_fontsize('16')
laser4.set_xlabel('Tiempo').set_fontsize('16')

show()
```
