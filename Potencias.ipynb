```python
#Importar librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from datetime import datetime
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
dataframe1=pd.read_csv('Files/Potencia660.csv')
dataframe2=pd.read_csv('Files/Potencia780.csv')
dataframe3=pd.read_csv('Files/Potencia808.csv')
dataframe4=pd.read_csv('Files/Potencia850.csv')
```

```python
corriente660 = dataframe1['Corriente']
potencia660 = dataframe1['Potencia']
print(corriente660)
```

```python
x = np.linspace(0,250,500)
def curva1(x):
    y1 = -0.1014285714+0.004285714286*x
    return y1 

def curva2(x):
    y2 = -31.5107604+0.8020086083*x
    return y2

y1 = curva1(x)
y2 = curva2(x)
print(len(y2))
print(len(x))
```

```python
plt.figure(figsize=(15,10))
plt.minorticks_on()
plt.grid(which='major',linestyle='-')
plt.grid(which='minor',linestyle=':')
plt.xlabel('Corriente (mA)', fontsize=20)
plt.ylabel('Potencia (mW)', fontsize=20)
plt.tick_params(axis='both', labelsize=15)
plt.ylim(0,90)
plt.xlim(0,190)

plt.title('Potencia vs. Corriente 660nm', fontsize=20)

plt.plot(x,y1, label = "y = -0.1014285714+0.004285714286*x")
plt.plot(x,y2, label = "y = -31.5107604+0.8020086083*x")
plt.plot(corriente660,potencia660, label = "Potencia vs. Corriente")
plt.legend(prop={'size': 12})
plt.show()

#Corriente Umbral
a = np.array([[-0.004285714286, 1],[-0.8020086083,1]])
b = np.array([-0.1014285714, -31.5107604])

print(np.linalg.solve(a,b))
```
